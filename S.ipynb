```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

%matplotlib inline
```

## 1. Análise Exploratória e Limpeza de Dados
[cite_start]Nesta etapa, realizamos o carregamento dos dados e o tratamento de valores nulos e inconsistências[cite: 30, 31]. [cite_start]Também exploramos a distribuição da variável alvo para entender o balanceamento das classes[cite: 32].

```python
# Carregamento e limpeza de colunas irrelevantes
df = pd.read_csv('Dataset_spine.csv').iloc[:, :13]

# Visualização da distribuição
df['Class_att'].value_counts().plot(kind='bar', color=['#3498db', '#e74c3c'])
plt.title('Distribuição: Abnormal vs Normal')
plt.show()
```

# Carregamento e limpeza de colunas irrelevantes
df = pd.read_csv('Dataset_spine.csv').iloc[:, :13]

# Visualização da distribuição
df['Class_att'].value_counts().plot(kind='bar', color=['#3498db', '#e74c3c'])
plt.title('Distribuição: Abnormal vs Normal')
plt.show()

```python
# Remoção de outliers usando o método IQR
for col in df.select_dtypes(include=np.number).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]

print(f"Registros finais após limpeza: {len(df)}")
```

## 2. Desenvolvimento de Modelos de Machine Learning
[cite_start]Realizamos a divisão dos dados em conjuntos de treino e teste (proporção 70/30) para validar o desempenho do modelo em dados não vistos[cite: 35]. [cite_start]Utilizamos o algoritmo **Random Forest** por sua robustez em tarefas de classificação[cite: 34].

```python
X = df.drop(columns=['Class_att'])
y = df['Class_att']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

modelo = RandomForestClassifier(random_state=42)
modelo.fit(X_train, y_train)
```

## 3. Avaliação e Importância das Variáveis
[cite_start]Abaixo, apresentamos as métricas de desempenho (Acurácia e F1-Score) e a Matriz de Confusão para validar a eficácia do diagnóstico preditivo[cite: 36, 37].

```python
y_pred = modelo.predict(X_test)
print(classification_report(y_test, y_pred))

ConfusionMatrixDisplay.from_estimator(modelo, X_test, y_test, cmap='Blues')
plt.grid(False)
plt.show()
```
